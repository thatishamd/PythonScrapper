```python
from bs4 import BeautifulSoup
import requests
```

```python
#getting the page as a document to make its object for Soup
Page = requests.get("http://www.arifbutt.me/")
PageContent = BeautifulSoup(Page.content,"html5lib")
```

```python
#YESSS finally succeeded in removing all useless a tags in p tag :D
tag = PageContent.p
PageContent.a.decompose()
print(tag)
```

```python
#MUAHAHA smooth, also removed all special charas
line = tag.get_text()
for word in line.split():
    if word.isalnum():
        print(word)
```
