```python
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python
#getting the page as a document to make its object for Soup
Page = requests.get("https://www.geo.tv/")
PageContent = BeautifulSoup(Page.content,"html5lib")
```

```python
#declaring all useable vars here ^^
uniqueVocabALLfiles = []
uniqueVocab = []
fname = "file"
vocabFName = "Unique_Vocab_file"
fcounter = 1
DictionaryNumToWord = {}
DictionaryWordToNum = {}
DictionaryCounter = 1
```

```python
#Some functions for later use
def searchInList(uniq,name):
    for item in uniq:
        if name in item:
            return uniq.index(item)
            break
    return -1

def allUnique(uniqueOfFile):
    for item in uniqueOfFile:
        if searchInList(uniqueVocabALLfiles,item[0]) == -1:
            uniqueVocabALLfiles.append(list((item)))
        else:
            index = searchInList(uniqueVocabALLfiles,item[0])
            uniqueVocabALLfiles[index][1]+=item[1]

def termDocument():
    #reading files and getting each words to make term document matrix
    matrix = np.zeros((fcounter - 1,len(uniqueVocabALLfiles)),dtype=np.int16)
    for i in range(1,fcounter):
        fileReader = open(fname + str(i) + ".txt")
        line = fileReader.read()
    for word in line.split():
        if(word.isalnum()):
            #print(word)
            index = searchInList(uniqueVocabALLfiles,word)
            if(index == -1):
                continue
            else:
                matrix[i-1][index] = 1;
    #print('\n')
    #Term Document Matrix Made successfully

def sortOnKey(val):
    return val[1]

def intro():
    #print("INTRO Method!")
    print("Scrapping Is Successful!!!")
    print("the URL 'www.geo.tv' has been Scrapped")
    print("You now have all the Headlines of the Present Time!\n")

def menu():
    print("What would you like to do? ")
    print("1) Display the whole Scrapped Content")
    print("2) Display the Unique Vocabulary of whole Content")
    print("3) Display the Content Dictionary of Words to Num Format")
    print("4) Display the Content Dictionary of Num to Words Format")
    print("5) Display the Term Document Matrix of all Scrapped Content")
    print("6) Search a News with words")
    print("7) Exit")
    
    #Users Choice
    try:
        choice = input("Enter You'r Choice: ")
        print(choice)
        if int(choice) < 1 and int(choice) > 7:
            raise Exception("Wrong Choice!")
        return int(choice)
    except:
        print("Something went Wrong")

def userQuery():
    #print("User Query Method")
    try:
        QueryLine = input("Enter the Words you want to search in ALL Scrapped Content (Must be space separated): ")
        QueryWords = []
        
        #checking the input words from Dictionary and make a list: QueryWords
        #print(test1)
        for word in QueryLine.split():
            #print(word + "test2") #test is positive
            for item in uniqueVocabALLfiles:
                if word in item:
                    #print("test3")
                    QueryWords.append(word)
                    #print("test4")
        if len(QueryWords) == 0:
            print("No Such Words Exist in the Content")
        else:
            print("Any Words which did not exist in Content has been removed!")
        print(QueryWords)
        
        #Making a list for all Files along with a counter integer
        AllFiles = []
        for i in range(1,fcounter):
            AllFiles.append(list((fname + str(i),0)))
        
        #Starting the Matching Process
        for word in QueryWords:
            for item in AllFiles:
                fileReader = open(item[0] + ".txt")
                for line in fileReader:
                    tempList = line.split()
                    if word in tempList:
                        item[1]+=1
                fileReader.close()
        
        #print(AllFiles)
        #Sorting Process based on [*][1] elements of sublists of AllFiles List
        AllFiles.sort(key = sortOnKey, reverse = True)
        print(AllFiles)
        
    except:
        print("Something Went Wrong!")

def driver():
    intro()
    
    while True:
        inp = menu()
    
        if inp == 1:
            # display the Scrapped Content
            print(PageContent.find_all('p'))
        elif inp == 2:
            # display the list uniqueVocabALLfiles
            print(uniqueVocabALLfiles)
        elif inp == 3:
            # display Dictionary 1st
            print(DictionaryWordToNum)
        elif inp == 4:
            # display Dictionary 2nd
            print(DictionaryNumToWord)
        elif inp == 5:
            # create the matrix first and then display it
            print("total files: " + str(fcounter - 1))
            termDocument()
            for x in range(0,12):
                for y in range(0,len(uniqueVocabALLfiles)):
                    print(matrix[x][y],end = ' ')
                print(' ')
        elif inp == 6:
            # Entertain the user's query
            userQuery()
        elif inp == 7:
            break
        else:
            print("invalid Choice!")
```

```python
#iterating on basis od all 'p' tags
for tag in PageContent.find_all('p'):
    
    #Writing the p tag content into a file
    fileWriter = open(fname + str(fcounter) + ".txt","w+")
    line = tag.get_text()
    line = line.casefold()
    #print(line)
    fileWriter.write(line)
    
    #checking each word of the content and making a list of Unique Vocabulary lists
    for word in line.split():
        if word.isalnum():
            #print(word)
            if searchInList(uniqueVocab,str(word)) == -1:
                uniqueVocab.append(list((word,1)))
            else:
                numberInList = searchInList(uniqueVocab,word)
                uniqueVocab[numberInList][1]+=1

    #print("\n")
    fileWriter.close()
    
    
    #Creating and writing Unique Vocabulary File for Each file created above ^^
    fileWriter = open(vocabFName + str(fcounter) + ".txt","w+")
    for item in uniqueVocab:
        for item in item:
            fileWriter.write(str(item))
            fileWriter.write(" ")
        fileWriter.write('\n')
    fileWriter.close()
    
    #removed code for individual unique Dictionaries and wrote new code for all Unique words Dictionary
    
    #Some changes for next Iteraton
    #print(uniqueVocab)
    AllUnique(uniqueVocab)
    uniqueVocab.clear()
    fcounter+=1
```

```python
#Making both Dictionaries here using the UniqueVocab List
for item in uniqueVocabALLfiles:
    DictionaryNumToWord[DictionaryCounter] = item[0]
    DictionaryWordToNum[item[0]] = DictionaryCounter
    DictionaryCounter+=1
    #print(item[0])
```

```python
#Main Segment
driver()
```
